```python
import matplotlib.pyplot as plt
```

```python
sales = [81.15, 73.66, 113.66, 157.43, 154.4, 196.48, 499.22, 561.07, 781.98]
year = [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023]

fig, ax = plt.subplots(figsize=(6, 4), dpi=200 )
plt.title('EU27 Vehicle imports from China in thousands', fontsize=14)

ax.bar(year, sales, color='darkblue')
ax.tick_params(axis='x', which='both', bottom=False, top=False, labelbottom=True)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

fig.text(0.05, 0.08, 'Figure 4: Car imports from China have surged since the pandemic with\n'
                    'EVs following a similar trajectory, demonstrating China\'s production\n'
                    'overcapacity.', 
         ha='left', va='center', fontsize=10)

fig.subplots_adjust(bottom=0.2)

plt.annotate('Source: Statista', xy=(0.5, 0.05), xycoords='figure fraction', ha='right', va='top', fontsize=10, color='gray')
plt.show()
```
